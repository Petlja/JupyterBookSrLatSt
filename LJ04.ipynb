## 4.1. Prosek niza

Postoji mnogo situacija kada se niz brojeva može zgodno opisati nekom _karakterističnom vrednošću_. Na primer, učenici dobijaju ocene iz mnogo predmeta, ali na kraju školske godine školski sistem želi da "izvede" jednu ocenu i da tom ocenom okarakteriše rad učenika u toj godini.

Vremenom se kao jedan od pogodnih predstavnika niza brojeva pokazao _prosek_. Prosek niza brojeva se računa ovako:

$$
  \hbox{prosek niza } [x_1, x_2, \dots, x_n] = \frac{x_1 + x_2 + \dots + x_n}{n}.
$$

Dakle, da bismo dobili prosek niza brojeva saberemo sve elemente niza i tako dobijeni broj podelimo dužinom niza.

U Pajtonu se obe ove aktivnosti mogu obaviti pozivom odgovarajuće ugrađene funkcije: funkcija `len` računa dužinu niza, dok
funkcija `sum` računa zbir elemenata niza brojeva.

Na primer:

```python
L = [1, 3, 5, 7, 9, 10]
print(len(L))
print(sum(L))
```

(Primeti da su ove funkcije _ugrađene u Pajton_ i da nema potrebe loviti ih po bibliotekama!)

Prema tome, prosek niza brojeva se u Pajtonu računa ovako:

```python
print("Prosek niza brojeva", L, "je", sum(L) / len(L))
```

Kako ćemo u mnogo navrata računati prosek, definisaćemo novu funkciju koja ga računa. U prethodnim razredima smo se već sreli sa definicijama novih funkcija u Pajtonu i zato mali program koji sledi ne bi trebalo da predstavlja iznenađenje za tebe:

```python
def prosek(L):
    return sum(L) / len(L)
```

Prethodni program je u sistem uneo novu definiciju: funkcija `prosek` će prihvatiti neki niz `L` i onda će kao rezultat svog rada vratiti zbir elemenata niza `L` podeljen dužinom niza `L` (dakle, prosek elemenata tog niza).

Primeti da je spisak Pajton naredbi koje predstavljaju _telo funkcije_ i objašnjavaju šta funkcija radi uvučen za četiri praznine.

Da vidimo kako radi naša nova funkcija:

```python
print("Prosek niza brojeva", L, "je", prosek(L))
```

U ćeliji ispod unete su ocene jednog učenika iz informatike i prikazan je njihov prosek:

```python
ocene = [5, 3, 5, 4, 4, 5]
print("Prosek ocena iz informatike je", prosek(ocene))
```

Ako ćemo iskreno, prosek ocena se uvek prikazuje zaokružen na dve decimale. U Pajtonu postoji ugrađena funkcija `round(x, n)` koja zaokružuje dati broj `x` na `n` decimala. Na primer:

```python
round(1.666666666, 2)
```

Prema tome, prosek ocena iz informatike sada možemo da dobijemo ovako:

```python
print("Prosek ocena iz informatike je", round(prosek(ocene), 2))
```

## 4.2. Sortiranje niza i medijana

Sortirati niz znači ispremeštati njegove elemente tako da budu poređani po veličini, od manjih ka većim ili obrnuto. Na primer:

$$
   [3, 1, 2, 5, 0, -1, 4] \to [-1, 0, 1, 2, 3, 4, 5]
$$

Ugrađena funkcija `sort` sortira niz i poziva se ovako:

```python
L = [3, 1, 2, 5, 0, -1, 4]
L.sort()
```

Ako proverimo vrednost promenljive `L` nakon poziva funkcije `sort` videćemo da su elementi liste sada poređani od manjih ka većim vrednostima:

```python
L
```

Ako želimo da poređamo elemente liste `L` od većih ka manjim vrednostima, to možemo da uradimo ovako:

```python
L.sort(reverse=True)
L
```

Opcija `reverse=True` kaže funkciji `sort` da želimo da sortiramo elemente liste u "obrnutom" poretku: od većih ka manjim vrednostima.

Ako su podaci sa kojima radimo "jako razbacani" i "imaju velika odstupanja" može se desiti da prosek niza _nije dobar predstavnik celog niza_. Zato se kao druga karakteristična vrednost niza često koristi _medijana_.

## 4.3. Filtriranje podataka

_Filtrirati podatke_ znači iz datog niza podataka izdvojiti one koji su nam na neki način interesantni.

Da bismo shvatili kako je moguće da je u prethodnom primeru prosečna zarada šest puta veća od medijalne, _filtriraćemo_ natprosečne zarade. Filtriranje liste u Pajtonu se može postići konstrukcijom koja izgleda ovako:

    [x for x in Lista if Uslov(x)]
    
Ova konstrukcija u novu listu pokupi sve elemente `x` liste `Lista` koji ispunjavaju `Uslov`. To je kao kad u matematici napišemo:

$$
  \{x \mid x \in \mathrm{Lista} \land \mathrm{Uslov}(x) \}
$$

Nama su u prethodnom primeru interesantne zarade koje su iznad proseka. Da bismo videli koliko ih ima i koje su to zarade prvo ćemo prosečnu zaradu smestiti u promenljivu i onda ćemo filtrirati listu `zarade`:

```python
prosecna_zarada = prosek(zarade)
natprosecne_zarade = [x for x in zarade if x >= prosecna_zarada]
```

Natprosečne zarade su:

```python
natprosecne_zarade
```

i vidimo da ih ima svega

```python
len(natprosecne_zarade)
```

Dakle, u ovoj kompaniji 12 zaposlenih ima ogromne zarade, a 10.000 zaposlenih ima zarade koje su ispod proseka. Ovih dvanaest nesrazmerno visokih zarada povećava prosek i daje utisak da "prosečna zarada u ovoj kokmpaniji i nije toliko loša", iako je stvarno stanje znatno lošije.

Situacije u kojima se medijana jako razlikuje od proseka nam govore da su podaci izuzetno neravnomerni i da je potrebno uložiti dodatni napor da bi se shvatilo kakvo stanje oni opisuju.

## 4.4. Prikazivanje podataka, njihovog proseka i njihove medijane

Jedan način da vidimo odnos proseka i medijane i kako oni predstavljaju podatke je da situaciju vizuelizujemo.

Krenimo od jednog jednostavnog primera. Pretpostavimo da je na polugodištu jedan učenik imao sledeće ocene iz navedenih predmeta:

| Predmet | Ocena |
|---------|-------|

| Srpski | 4 |
| Likovno | 5 |

| Fizičko | 5 |
| Muzičko | 4 |

Podatke možemo predstaviti pomoću dve liste, ovako:

```python
predmeti = ["mat", "srp", "lik", "ist", "fiz", "muz", "tio"]
ocene    = [2,     4,     5,     3,     5,     4,     5    ]
```

Želimo da prikažemo njegove ocene histogramom i da prosečna ocena bude označena crvenom linijom kako bismo mogli da vidimo koje ocene su iznad proseka, a koje ispod.

Prvo ćemo učitati biblioteku za crtanje grafikona:

```python
import matplotlib.pyplot as plt
```

Histogram sa ocenama možemo da nacrtamo veoma jednostavno:

```python
plt.figure(figsize=(10,5))
plt.bar(predmeti, ocene)
plt.title("Ocene na polugodistu")
plt.show()
plt.close()
```

Sada želimo da preko histograma nacrtamo crvenu liniju koja predstavlja prosečnu ocenu. Prvo ćemo izračunati prosečnu ocenu `po`:

```python
po = prosek(ocene)
```

a onda ćemo na gornji dijagram dodati i niz podataka koji izgleda ovako:

`[po, po, po, po, po, po, po]`

```python
plt.figure(figsize=(10,5))
plt.bar(predmeti, ocene)
plt.plot(predmeti, [po, po, po, po, po, po, po], color="r")
plt.title("Ocene na polugodistu")
plt.show()
plt.close()
```

Ako jedan isti podatak `s` treba ponoviti `n` puta možemo da pišemo kao gore:

`[s, s, s, (i tako dalje)]`

ali možemo da to zapišemo kraće ovako:

`[s]*n`

Oznakom `[s]*n` smo Pajtonu rekli da želimo da napravimo novi niz koji se dobija tako što se niz `[s]` "umnoži" `n` puta. Na primer, `[5]*7 = [5, 5, 5, 5, 5, 5, 5]`. (Takođe je `[6, 7, 8]*3 = [6, 7, 8, 6, 7, 8, 6, 7, 8]`, ali nam ovo za sada neće biti potrebno.)

Tako dobijamo sledeću verziju programa:

```python
plt.figure(figsize=(10,5))
plt.bar(predmeti, ocene)
plt.plot(predmeti, [po] * len(ocene), color="r")
plt.title("Ocene na polugodistu")
plt.show()
plt.close()
```

Za kraj ćemo dijagramu dodati legendu:

```python
plt.figure(figsize=(10,5))
plt.bar(predmeti, ocene, label="ocene")
plt.plot(predmeti, [po] * len(ocene), color="r", label="prosek")
plt.title("Ocene na polugodistu")
plt.legend()
plt.show()
plt.close()
```

Sada ćemo da vizuelizujemo zarade u kompaniji o kojoj smo govorili u prethodnim odeljcima. _(Lista sa kojom radimo je veoma velika, tako da ovo može potrajati! Samo hrabro o strpljivo!)_

```python
n = len(zarade)
plt.bar(range(n), zarade)
plt.show()
plt.close()
```

Pošto mi već znamo da je prosečna zarada oko 3912 američkih dolara i da ima samo 12 zaposlenih u toj kompaniji koji primaju toliko ili više, ograničićemo vrednosti koje se prikazuju na $y$-osi. Odmah ćemo prikazati i prosek crvenom linijom.

```python
n = len(zarade)
plt.figure(figsize=(15,5))
plt.bar(range(n), zarade, color="b")
plt.plot([0, n-1], [prosecna_zarada, prosecna_zarada], color="r")
plt.ylim(0,6000)
plt.show()
plt.close()
```

Na ovom dijagramu se jasno vidi kako su nesrazmerno velike zarade nekolicine zaposlenih "povukle prosek za sobom".

Pogledajmo sada gde je medijana. Nju ćemo predstaviti kao dijagram sa samo jednim crvenim stubićem na sredini niza.

```python
n = len(zarade)
plt.figure(figsize=(15,5))
plt.bar(range(n), zarade, color="b")
plt.bar([n//2], [zarade[n//2]], color="r")
plt.ylim(0,6000)
plt.show()
plt.close()
```

Gde je crvena crtica?

Pošto na relativno malom prostoru prikazujemo preko 10.000 podataka pojedinačni stubići su jako uski i stubić koji predstavlja medijana se ne vidi. Da bismo je istakli promenićemo širinu stubića opcijom `width`.

```python
n = len(zarade)
plt.figure(figsize=(15,5))
plt.bar(range(n), zarade, color="b")
plt.bar([n//2], [zarade[n//2]], color="r", width=100)
plt.ylim(0,6000)
plt.show()
plt.close()
```
